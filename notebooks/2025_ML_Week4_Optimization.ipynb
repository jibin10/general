    0: Setosa (Orange)

#### More saturated colors $=>$ lower gini score (meaning samples are mostly from one class)
#### Lighter colors $=>$ higher gini score (samples are more mixed among classes)


### What is the Gini score?
#### It's related to an economic inequality indicator which ranges from 0 (equitable) to 1 (inequitable)
#### The lower the better -> Meaning the variance is smaller



#### There's also another indicator called Entropy



#### If the samples mostly belong to one class, the entropy would be lower (ex: box has only 100 yellow balls). 
#### Otherwise, entropy would be higher (ex: box has 50 yellow balls and 50 green balls).

### The higher the Information Gain the better

## Confused?
#### In summary,
##### Gini: Lower the better 
##### Entropy: Lower the better
##### Information Gain: Higher the better

## How to check for overfitting

### What is a Decision Boundary?
#### A decision boundary is the region of a problem space in which the output label of a classifier is ambiguous. If the decision surface is a hyperplane, then the classification problem is linear, and the classes are linearly separable. Decision boundaries are not always clear cut.

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("3 Class values with 2 Features Sample data creation")

# Using 2 features, 3 classes, and making into 2 dimensions
X_features, y_labels = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_classes=3, n_clusters_per_class=1,random_state=0)

# using scatter plot with 2 features in 2 dimensions and label color differently
plt.scatter(X_features[:, 0], X_features[:, 1], marker='o', c=y_labels, s=25, cmap='rainbow', edgecolor='k')
plt.show()
```

```python
import numpy as np

# Visualize Decision Boundary
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # Create Scatter Plot
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # Train
    model.fit(X, y)
    # msehgrid format
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # visualization using contours
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Plot
dt_clf = DecisionTreeClassifier().fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
plt.show()
```

```python
# Recall that we can tune one of the parameters:
# min_samples_leaf=6 
dt_clf = DecisionTreeClassifier( min_samples_leaf=6).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
plt.show()
```

```python
dt_clf = DecisionTreeClassifier( min_samples_leaf=5).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
plt.show()
```

```python
dt_clf = DecisionTreeClassifier( min_samples_leaf=1).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
plt.show()
```
